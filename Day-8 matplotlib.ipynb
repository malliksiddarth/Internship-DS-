```python
normal.cumsum().plot(grid=True)
```

```python
winratio = pd.DataFrame({'CSK': np.random.normal(size=10), 
                       'SRH': np.random.gamma(1, size=10), 
                       'RCB': np.random.poisson(size=10)})
print(winratio)

winratio.cumsum(0).plot()
```

```python
winratio.cumsum(0).plot(subplots=True)
```

```python
winratio.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,win in enumerate(['CSK','SRH','RCB']):
    winratio[win].cumsum(0).plot(ax=axes[i], title=win)
axes[0].set_ylabel('cumulative sum')
```

```python
rng = np.random.RandomState(7)
x = rng.randn(60)
y = rng.randn(60)
colors = rng.rand(60)
sizes = 1000 * rng.rand(60)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.8,
            cmap='Blues')
plt.colorbar()  
```

```python
x = np.linspace(0, 10, 20)
dy = 0.5
y = np.sin(x) + dy * np.random.randn(20)

plt.errorbar(x, y, yerr=dy, fmt='b');
```
